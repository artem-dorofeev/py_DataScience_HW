*Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"*

```python
position = df.columns.get_loc('Мова програмування')
df_cleaned = df.dropna(axis=1, how='any')
df_cleaned.insert(position, 'Мова програмування', df['Мова програмування'])
df_cleaned.head()
```

*Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"*

```python
missing_values = df_cleaned.isnull().sum()

total_values = len(df)
missing_percentage = (missing_values / total_values) * 100

missing_data = pd.DataFrame({
    'Кількість пропусків': missing_values,
    'Частка пропусків (%)': missing_percentage
})

missing_data
```

*Видаліть усі рядки у вихідній таблиці за допомогою методу dropna*

```python
df_next = df_cleaned.dropna()
df_next
```

*Визначте новий розмір таблиці за допомогою методу shape*

```python
df_next.shape
```

*Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python*

```python
python_data = df_next[df_next['Мова програмування'] == 'Python']
python_data
```

*Визначте розмір таблиці python_data за допомогою методу shape*

```python
python_data.shape
```

*Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"*

```python
grouped_data = python_data.groupby('Посада')
grouped_data
```

*Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"*

```python
aggregated_data = grouped_data['Зарплата на місяць'].agg(['min', 'max'])

# aggregated_data = grouped_data.agg({
#     'Зарплата.в.месяц': ['min', 'max']
# }).reset_index()

aggregated_data
```

*Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.*

```python

def fill_avg_salary(dataframe):
    return dataframe['Зарплата на місяць'].mean()

aggregated_data['avg'] = grouped_data.apply(fill_avg_salary)
aggregated_data
```

*Створіть описову статистику за допомогою методу describe для нового стовпчика.*

```python
avg_statistics = aggregated_data["avg"].describe()
avg_statistics
```

*Збережіть отриману таблицю в CSV файл*

```python
aggregated_data.to_csv('/content/agg_salary.csv')
avg_statistics.to_csv("/content/avg_statistic.csv")
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.countplot(x='Мова програмування', data=df_cleaned)
plt.xticks(rotation=45, ha='right')
plt.title('Кількість респондентів за мовою програмування')
plt.xlabel('Мова програмування')
plt.ylabel('Кількість респондентів')
plt.show()
```

```python
df_python = df_cleaned[df_cleaned['Мова програмування'] == 'Python']

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Досвід роботи на поточному місці', y='Зарплата на місяць', data=df_python, s=200)
plt.title('Залежність зарплати від досвіду роботи для мови програмування Python')
plt.xlabel('Досвід роботи на поточному місці')
plt.ylabel('Зарплата на місяць')
plt.grid(True)
plt.show()
```
