```python
agg_quantity = {
    'name': 'count'
}
df_quantity_books = df.groupby('author').agg(agg_quantity)
# df_quantity_books.columns = ['quantity']
df_quantity_books = df_quantity_books.rename(columns={'name': 'quantity'})

df_quantity_books
```

*Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні*

```python
agg_mean_rating = {
    'user_rating': 'mean'
}

result_mean_rating = df.groupby('author').agg(agg_mean_rating)
result_mean_rating.columns = ['average_rating']

result_mean_rating
```

```python
min_mean_author = result_mean_rating['average_rating'].idxmin()
min_mean_rating = result_mean_rating.loc[min_mean_author, 'average_rating']

print("Автор:", min_mean_author)
print("Середній рейтинг:", min_mean_rating)
```

*У якого автора середній рейтинг мінімальний?* - ***Donna Tartt***

*Який у цього автора середній рейтинг?* - ***3.9***

*З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну*

```python
merged_df = pd.concat([result_mean_rating, df_quantity_books], axis=1)

merged_df
```

*Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)*

```python
sorted_df = merged_df.sort_values(by=['quantity', 'average_rating'], ascending=[True, True])

sorted_df
```
